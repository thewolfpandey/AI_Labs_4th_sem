```python
#creating a Spectrogram out of my cat_meme.mp3 file...
import scipy.io.wavfile as wav
from librosa import load, stft
import librosa
from librosa import display
```

```python
audio_file_path = 'cat_meme.mp3'
y,sr = librosa.load(audio_file_path)
```

```python
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

plt.figure(figsize=(10, 6))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```

```python
pdf_file_path = 'lab1_ai.pdf'

with pdfplumber.open(pdf_file_path) as pdf:
    for page_number in range(len(pdf.pages)):
        page = pdf.pages[page_number]
        text = page.extract_text()
        
        print(f"Page {page_number + 1}:\n{text}\n")
```

```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
mha_file_path = 'mha_image.mha'
image = sitk.ReadImage(mha_file_path)
image_array = sitk.GetArrayFromImage(image)

slice_index = 50
plt.imshow(image_array[slice_index, :, :], cmap='gray')
plt.title(f'3D MHA Image - Slice {slice_index}')
plt.show()
```
