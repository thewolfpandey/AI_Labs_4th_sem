```python
#NAME:- ANMOL PANDEY
#Enrolment Number:- E22CSEU1069
#Batch:- 36

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
```

```python
print('Shape of the dataset is:- ', df.shape)
df.head()
```

```python
#Q1
total_profit = df["total_profit"]
months = df["month_number"]

plt.figure(figsize=(5,3))

plt.plot(months, total_profit)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Total Profit of All Months")
plt.show()
```

```python
#Q2
plt.figure(figsize=(5,3))

plt.plot(months, df["total_units"], linestyle='dotted', color='blue', marker='o', markerfacecolor='red', markersize=8, label='Sold units number')
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(loc='lower right')
plt.title("Total Profit with Style Properties")
plt.show()
```

```python
#Q3
plt.figure(figsize=(6,4))

products = df.columns[1:-2]
for product in products:
    plt.plot(months, df[product], label=product)

plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.legend()
plt.title("Number of Units Sold per Month for Each Product")
plt.show()
```

```python
#Q4
plt.figure(figsize=(6,4))
plt.scatter(months, df["toothpaste"], marker='o')
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.title("Toothpaste Sales Data with Grid")
plt.grid(True)
plt.show()
```

```python
#Q5
reqd_data = df[["facecream", "facewash"]]
reqd_data.plot(kind='bar', stacked=True)
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.title("Face Cream and Facewash Sales Data")
plt.show()
```

```python
#Q6
bathing_soap_data = df["bathingsoap"]
plt.bar(months, bathing_soap_data)
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.title("Bathing Soap Sales Data")
plt.savefig("soap_sales.png")
```

```python
#Q7
plt.hist(df["total_profit"], bins=10, edgecolor='black')
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.title("Histogram of Total Profit")
plt.show()
```

```python
#Q8
total_units_per_product = (df.drop(['total_profit', 'total_units', 'month_number'], axis=1)).sum(numeric_only=True)
print(total_units_per_product)

plt.figure(figsize=(10, 6))
plt.pie(total_units_per_product, labels=total_units_per_product.index, autopct='%1.1f%%')
plt.title("Total Units Sold per Year for Each Product (Pie Chart)")
plt.show()
```

```python
#Q9
fig, axs = plt.subplots(2, 1)
axs[0].bar(months, df["bathingsoap"])
axs[0].set_title("Bathing Soap Sales Data")
axs[1].bar(months, df["facewash"])
axs[1].set_title("Facewash Sales Data")
plt.tight_layout()
plt.show()
```

```python
#Q10
df[products].plot(kind='area', stacked=True)
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.title("Stack Plot for All Product Sales Data")
plt.show()
```
