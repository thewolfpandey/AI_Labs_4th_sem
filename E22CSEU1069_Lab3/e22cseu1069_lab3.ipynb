```python
exp = np.array([16, 12, 18, 4, 3, 10, 5, 12])
per_rating = np.array([87, 88, 89, 68, 78, 80, 75, 83])
```

```python
slope, intercept = np.polyfit(exp, per_rating, 1)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# plt.figure(figsize=(3,2))
plt.scatter(exp, per_rating, label='Data Points')
plt.show()
```

```python
x_vals = np.linspace(min(exp), max(exp), 100)
y_vals = slope*x_vals+intercept
plt.scatter(exp, per_rating, label='Data Points')
plt.plot(x_vals, y_vals, color='red', label='Best-Fitted Line')
plt.xlabel('Experience (in years)')
plt.ylabel('Performance Rating')
plt.legend()
```

```python
estimated_rating_20_years = slope * 20 + intercept
print(f"Estimated performance rating for 20 years of experience: {estimated_rating_20_years}")
```

```python
#Quest.2

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.utils import shuffle
```

```python
bh_df = pd.read_csv('BostonHousing.csv')
print("Shape:- ", bh_df.shape)
bh_df.head()
```

```python
bh_X = bh_df.drop(['medv'], axis=1)
print("Shape of my X:- ", bh_X.shape)
bh_X.head()
```

```python
bh_Y = bh_df['medv']
print("Shape of my Y:- ", bh_Y.shape)
bh_Y.head()
```

```python
bh_X_train, bh_X_test, bh_Y_train, bh_Y_test = train_test_split(bh_X, bh_Y, test_size=0.2, random_state=69)
print(bh_X_train.shape)
print(bh_X_test.shape)
print(bh_Y_train.shape)
print(bh_Y_test.shape)
```

```python
bh_model = LinearRegression()
bh_model.fit(bh_X_train, bh_Y_train)

bh_Y_pred = bh_model.predict(bh_X_test)
bh_mse = mean_squared_error(bh_Y_test, bh_Y_pred)
bh_rmse = mean_squared_error(bh_Y_test, bh_Y_pred, squared=False)
bh_mae = mean_absolute_error(bh_Y_test, bh_Y_pred)
bh_r2 = r2_score(bh_Y_test, bh_Y_pred)

print("Mean Squared Error:", bh_mse)
print("Root Mean Squared Error:", bh_rmse)
print("Mean Absolute Error:", bh_mae)
print("R-squared Score:", bh_r2)
```

```python
splits = [0.6, 0.5, 0.7, 0.8, 0.9, 0.95]
for split in splits:
    bh_X_train, bh_X_test, bh_Y_train, bh_Y_test = train_test_split(bh_X, bh_Y, test_size=1-split, random_state=42)
    bh_model.fit(bh_X_train, bh_Y_train)
    bh_Y_pred = bh_model.predict(bh_X_test)
    bh_mse = mean_squared_error(bh_Y_test, bh_Y_pred)
    print(f"\nTrain-Test Split: {split * 100}-{(1 - split) * 100}")
    print("Mean Squared Error:", bh_mse)

seeds = [42, 123, 456]
for seed in seeds:
    bh_X_train, bh_X_test, bh_Y_train, bh_Y_test = train_test_split(bh_X, bh_Y, test_size=0.2, random_state=seed)
    bh_model.fit(bh_X_train, bh_Y_train)
    bh_Y_pred = bh_model.predict(bh_X_test)
    bh_mse = mean_squared_error(bh_Y_test, bh_Y_pred)
    print(f"\nRandom Seed: {seed}")
    print("Mean Squared Error:", bh_mse)
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
def compute_cost(x, y, w, b): 
    m = x.shape[0] 
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
def compute_gradient(x, y, w, b): 
    m = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb = w * x[i] + b 
        dj_dw_i = (f_wb - y[i]) * x[i] 
        dj_db_i = f_wb - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
        
    return dj_dw, dj_db
```
