```python
df = np.array([[1, 2.4], [2, 4.5], [3, 6.1], [4, 8.0], [5, 10.2], [6, 12.5], [7, 15.1], [8, 18.0], [9, 21.2], [10, 24.5]])
```

```python
X = df[:, 0]
y = df[:, 1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train.reshape(-1, 1), y_train)
```

```python
y_pred_lin = lin_reg.predict(X_test.reshape(-1, 1))
mse_lin = np.mean((y_test - y_pred_lin)**2)
r2_lin = lin_reg.score(X_test.reshape(-1, 1), y_test)
```

```python
print("L.R:")
print(f"MSE:->{mse_lin:.4f}")
print(f"R-2:->{r2_lin:.4f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
models = {}
X_train=X_train.reshape(-1, 1)
X_test=X_test.reshape(-1, 1)
y_train=y_train.reshape(-1, 1)

for degree in range(2, 5):
    poly_reg = PolynomialFeatures(degree=degree)
    X_train_poly = poly_reg.fit_transform(X_train)
    poly_reg_model = LinearRegression()
    poly_reg_model.fit(X_train_poly, y_train)
    y_pred_poly = poly_reg_model.predict(poly_reg.transform(X_test))
    mse_poly = np.mean((y_test - y_pred_poly)**2)
    r2_poly = poly_reg_model.score(poly_reg.fit_transform(X_test), y_test)
    models[f"P.R (deg {degree})"] = {
        "mse": mse_poly,
        "r2": r2_poly
    }
    plt.figure(figsize=(4, 3))
    plt.scatter(X_test, y_test, color='blue', label='Actual')
    X_plot = np.linspace(0, 11, 100).reshape(-1, 1)
    X_plot_poly = poly_reg.transform(X_plot)
    y_plot = poly_reg_model.predict(X_plot_poly)
    plt.plot(X_plot, y_plot, color='red', label=f'Degree {degree}')
    plt.plot(X_test, y_pred_lin, color='red', label='Linear')

    
    plt.title(f'Polynomial Regression (Degree {degree})')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
for model_name, metrics in models.items():
    print(f"{model_name}:")
    print(f"MSE: {metrics['mse']:.4f}")
    print(f"R2: {metrics['r2']:.4f}\n")
```

```python
import pandas as pd
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
# Data ka ID column model ke training mei use nahi hoga(bcoz, usse decide nahi hoga ki cancer, Hai ya nahi!!)
df.drop("id", axis=1, inplace=True)
```

```python
df.head(1)
```

```python
X = df.drop('diagnosis', axis=1)
Y = df['diagnosis'].replace({'M': 1, 'B': 0})
Y.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
y_pred.shape
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = [4, 3] 

conf = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots()
sns.heatmap(conf, ax=ax, annot=True)
ax.set_title("Logistic Reg ka Confusion Matrix")
ax.set_xlabel('Predicted label')
ax.set_ylabel('True label')
plt.show()
```

```python
precision, recall, _, _ = precision_recall_fscore_support(y_test, y_pred)
print(precision)
print("----------------------")
print(recall)
```
